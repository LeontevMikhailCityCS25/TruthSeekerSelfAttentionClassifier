```python
import numpy as np
from SelfAttention import SingleHeadSelfAttention, MultiHeadSelfAttention
```

```python
single = SingleHeadSelfAttention(5)
vector = np.random.rand(10, 5)
print(vector.shape)
print(vector)
```

```python
output = single.forward(vector)
print(output.shape)
print(output)
```

```python
print("Q\n",single.W_q)
print("K\n",single.W_k)
print("V\n",single.W_v)
```

```python
for i in range(output.shape[0]):
    print(f"Output vector {i}: {output[i]}")
    print(f"Input vector {i}: {vector[i]}")
```

```python
single.backward(d_out=np.ones_like(output), X=output ,learning_rate=0.01)
```

```python
print("Q\n",single.W_q)
print("K\n",single.W_k)
print("V\n",single.W_v)
```
